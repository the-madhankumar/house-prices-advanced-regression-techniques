```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_2 = pd.read_csv(r"D:\projects\ML\house-prices-advanced-regression-techniques\test.csv")
df_1 = pd.read_csv(r"D:\projects\ML\house-prices-advanced-regression-techniques\train.csv")
if 'SalePrice' not in df_2.columns:
    df_2["SalePrice"] = 0
df = pd.concat([df_1,df_2], axis = 0)
df = df.set_index('Id')
```

```python
df_objects = df[df.select_dtypes(include=['object']).columns]
df = df.drop(df[df_objects.isnull().sum()[df_objects.isnull().sum() > 1100].index], axis = 1)
```

```python
df_objects = df_objects.drop(df_objects[df_objects.isnull().sum()[df_objects.isnull().sum() > 1100].index], axis = 1)
df_objects = df_objects.fillna('null')
df_objects_encoded = pd.get_dummies(df_objects)
```

```python
for i in df_objects_encoded:
    if 'null' in i:
        df_objects_encoded = df_objects_encoded.drop(i, axis = 1)
        print(i)
```

```python
new_df = pd.concat([df,df_objects_encoded], axis = 1)
```

```python
len(new_df.columns)
```

```python
new_df = new_df.drop(new_df.select_dtypes(include='object'), axis=1)
```

```python
new_df.isnull().sum()[new_df.isnull().sum() > 0]
```

```python
Mode_columns = ['GarageCars', 'GarageYrBlt', 'BsmtFullBath', 'BsmtHalfBath']
Mean_columns = ['LotFrontage', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'GarageArea']

for i in Mode_columns:
    new_df[i] = new_df[i].fillna(new_df[i].mode()[0])

for j in Mean_columns:
    new_df[j] = new_df[j].fillna(np.round(new_df[j].mean()))
    
new_df.isnull().sum()
```

```python
training_data = new_df[0:len(df_1)]
test_data = new_df[len(df_1):]
test_data = test_data.drop(columns = 'SalePrice')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
X = training_data.drop(columns='SalePrice')
y = training_data['SalePrice']
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2)
Y_train = np.reshape(Y_train,(-1,1))
Y_test = np.reshape(Y_test,(-1, 1))
X_train.shape, Y_train.shape
```

```python
model_1 = LinearRegression()
model_1.fit(X_train, Y_train)
y_pred = model_1.predict(X_test)
mean_squared_error(Y_test, y_pred)
```

```python
model_2 = XGBRegressor(n_estimators = 1000, learning_rate = 0.1, random_state = 42)
model_2.fit(X_train, Y_train)
y_pred = model_2.predict(X_test)
mean_squared_error(Y_test, y_pred)
```

```python
model_3 = RandomForestRegressor(n_estimators = 1000, random_state = 42)
model_3.fit(X_train, Y_train)
y_pred = model_3.predict(X_test)
mean_squared_error(Y_test, y_pred)
```

```python
sort_indices = np.argsort(Y_test)

plt.figure(figsize=(10, 10))
plt.plot(np.arange(len(Y_test)), Y_test, label='original')
plt.plot(np.arange(len(Y_test)), y_pred, label='Predicted')
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Predicted vs Original')
plt.legend()
plt.show()
```

```python
pred = model_2.predict(test_data)
final = pd.DataFrame()
final['Id'] = test_data.index
final['SalePrice'] = pred

final.to_csv('output.csv',index = False)
```
